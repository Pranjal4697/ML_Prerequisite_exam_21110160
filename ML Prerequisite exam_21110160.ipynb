**Solution for Q1**

Lets assume matrix A has dimensions (n,k) and B has dimension (k,m)

As we know that in Matrix Multiplication, elements of rows of one matrix get multiplied with corresponding elments of columns of another matrix to produce the output matrix.

Here in this problem, k elements of each row get multiplied with k elements of each columns of B so for each row of A , total multiplications will be mk.
Since there are total of n rows in A so the total number of **Multiplication**
involved in the operation will be **mkn**.

Similarly in each row column product there will be k-1 additions.Therefore,
total Number of **additions** will be **(k-1)mn**

**Solution for Q5**

```python
import numpy as np
import jax
import jax.numpy as jnp


def myGradient(x):
   
    return 2*x[0]*x[1] + (x[1]**3)*jnp.cos(x[0]), (x[0]**2) + 3*(x[1]**2)*jnp.sin(x[0])

def myFunction(x):
    return (x[0]**2)*x[1]+(x[1]**3)*jnp.sin(x[0])



jaxGradient = jax.grad(myFunction)
a=np.array([2.0,3.1])

print("JAX Gradient :", jaxGradient(a))
print("Calculated Expression Gradient: ", myGradient(a))
```

we can see that the Gradient Calculated using JAX and the mannually calculated function is same

**Solution for Q8**

```python
class subject:
    def __init__(self,subject1,Mark):
        self.subject=subject1
        self.Mark=Mark

class student:
    def __init__(self,RollNumber,Name,Subject_list=[]):
        self.RollNumber=RollNumber
        self.Name=Name
        self.Subjects=Subject_list
    def Add_subject(self,subject):
        self.Subjects.append(subject)

class branch:
    def __init__(self,branch_name,students_list=[]):
        self.branch_name=branch_name
        self.students=students_list
    def Add_student(self,student):
        self.student.append(student)
    
class year:
    def __init__(self,year,branch_list=[]):
        self.year=year
        self.branches=branch_list
    def Add_branch (self,branch):
        self.branches.append(branch)
        

Maths=subject("Maths",100)
English=subject('English','70')
Chemistry=subject('Chemistry','80')

student1=student('1','N',[Maths,English,Chemistry])
student2=student('2','M',[Maths,English,Chemistry])

branch1=branch('CSE',[student1])
branch2=branch('EE',[student2])

year1=year('2022',[branch1,branch2])

print('year1:',year1.year)
print(' Branch1:',year1.branches[0].branch_name)
print('  Student1:')
print('   RollNumber:',year1.branches[0].students[0].RollNumber)
print('   Name:',year1.branches[0].students[0].Name)
print('   Marks:')
print('    ',year1.branches[0].students[0].Subjects[0].subject,year1.branches[0].students[0].Subjects[0].Mark)
print('    ',year1.branches[0].students[0].Subjects[1].subject,year1.branches[0].students[0].Subjects[1].Mark)
        
```

**Solution for Q12**

Broadcasting in numpy describes how numpy arrays of different dimensions are treated in case of arithmetic operations
For example:

```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([3])
arr3=arr1*arr2

print ("arr3:",arr3)
```

In the above code we can see that the dimensions of arr1(1,3) is not same as arr2(1,1) but after multiplication we are getting each element of arr1 multiplied with 3,here 3 is broadcasted to the all the elements of arr1.

**Solution for Q13**

```python
import numpy as np
def My_argmin(arr):
    Min_Value=arr[0]
    Min_Index=0
    for i in range(0,len(arr)):
        if arr[i]<Min_Value:
            Min_Value=arr[i]
            Min_Index=i
    return Min_Index

arr=np.array([2,4,-3])
Min_Index=My_argmin(arr)
Built_In_function_Output=np.argmin(arr)

print("My_argmin Output: ",Min_Index)
print("Built In Function (np.argmin) output: ",Built_In_function_Output)
```
